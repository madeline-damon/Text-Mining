Okay, this is a mess. The first row has all the dates, the subsequent rows have headlines with many blanks put in. Let's start by getting all the dates, which are on the first row.

```python
##Beginning with date cleaning
###Removing just the dates from butte.txt
with open(working_dir + input_file) as infile :
    dates = infile.readline().strip().split("\t")

##Check to see that dates were pulled properly

print(dates)
```

```python
###Create an empty list for new dates
new_dates=[]

##How to change date format
for date in dates : ##For each item in list of dates
    day, month = date.split("-") ###There are two objects split by a hyphen, the day and month
    
    year = "2016" ##establishes 2016 as year once the loop arrives at January: Sept-Dec are manually changed to 2015
    
    if month == "Sep": ##If month is Sep, change Sep to 09
        new_month = "09"
        
        if int(day) < 22 : ##If day has an integer value of less than 22, assign year 2016
            year = "2016"
        else : 
            year = "2015"  ##Otherwise, assign the year 2015 to the date
            
    elif month == "Oct":  ##If month is Oct, change Oct to 10 and assign year 2015
        new_month = "10"
        year = "2015"
    elif month == "Nov":  ##If month is Nov, change Nov to 11 and assign year 2015
        new_month = "11"
        year = "2015"
    elif month == "Dec": ##If month is Dec, change Dec to 12 and assign year 2015
        new_month = "12"
        year = "2015"
    elif month == "Jan": ##If month is Jan, change Jan to 01 (keeps year assigned above of 2016)
        new_month = "01"
    elif month == "Feb":
        new_month = "02"
    elif month == "Mar":
        new_month = "03"
    elif month == "Apr":
        new_month = "04"
    elif month == "May":
        new_month = "05"
    elif month == "Jun":
        new_month = "06"
    elif month == "Jul":
        new_month = "07"
    elif month == "Aug":
        new_month = "08"
    else :
        new_month = "trouble"
        
    if len(day) ==1:    ###If the length of string day is only 1 character, add a zero before the string value
        day = "0" + day
    
    new_dates.append(year + "-" + new_month + "-" + day)   ##In the empty new_dates list, put objects together as year-new_month-day

    
```

```python
##Checking date format as I go to make sure if statement is working
new_dates

##Once all months have been added to the else if statement, all dates look correct
##Months have been assigned proper year with Sep thru Dec as 2015 and Jan-Sep as 2016
```
